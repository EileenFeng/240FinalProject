```gas
.ORIG x3000
.SET MODE TERMINAL
ST  R7, REG7
JSR PUSH 
AND R1, R1, 0
ADD R1, R1, 13
ST R1, NUM
ADD R1, R1, #-5
JSR ARRAY_UPDATE
JSR POP
LD  R7, REG7
HALT

PUSH:  ST R6, REG6     ;; save R5, R6 to MEMORY
       ST R5, REG5     ;; R7 has return address
                       ;; REG7 is in MEMORY already
                       ;; R6 and R5 are destroyed
       LD R6, SREG     ;; R6 is now TOP of stack
       STR R0, R6, #0      ;; push R0 on stack
       ADD R6, R6, #1  ;; incr
       STR R1, R6, #0      ;; push R1 on stack
       ADD R6, R6, #1  ;; incr
       STR R2, R6, #0      ;; push R2 on stack
       ADD R6, R6, #1  ;; incr
       STR R3, R6, #0      ;; push R3 on stack
       ADD R6, R6, #1  ;; incr
       STR R4, R6, #0      ;; push R4 on stack
       ADD R6, R6, #1  ;; incr
       STR R5, R6, #0      ;; push R5 on stack
       ADD R6, R6, #1  ;; incr
       LD R5, REG6     ;; get R6 from MEMORY (destroys R5)
       STR R5, R6, #0      ;; push R6 on stack
       ADD R6, R6, #1  ;; incr
       LD R5, REG7     ;; get R7 from MEMORY
       STR R5, R6, #0  ;; push R7 on stack
       ADD R6, R6, #1  ;; incr
       ST R6, SREG      ;; save incremented SREG
       LD R6, REG6
       LD R5, REG5     ;;
       RET
      
POP:   LD R6, SREG     ;; R6 is TOP of stack
       ADD R6, R6, #-1     ;; decr
       LDR  R1, R6, #0      ;; load R7
       ST  R1, REG7    ;; save to MEMORY
       ADD R6, R6, #-1     ;; decr
       LDR  R1, R6, #0      ;; load R6
       ST  R1, REG6    ;; save to MEMORY
       ADD R6, R6, #-1     ;; decr
       LDR  R5, R6, #0      ;; load R5
       ADD R6, R6, #-1     ;; decr
       LDR  R4, R6, #0      ;; load R4
       ADD R6, R6, #-1     ;; decr
       LDR  R3, R6, #0      ;; load R3
       ADD R6, R6, #-1     ;; decr
       LDR  R2, R6, #0      ;; load R2
       ADD R6, R6, #-1     ;; decr
       LDR  R1, R6, #0      ;; load R1
       ADD R6, R6, #-1     ;; decr
       LDR  R0, R6, #0      ;; load R0
       ST R6, SREG      ;; save decremented TOP
       LD R6, REG6, #0     ;; put R6 back the way it was
       RET
       
       
REG:    .BLKW    #8
SREG:   .FILL REG
REG5:   .FILL    #0
REG6:   .FILL    #0
REG7:   .FILL    #0
HALT

;;;;;;;;;adds number of image-parts to be revealed to array
ARRAY_UPDATE: JSR PROG
              JSR NINE_MULT
              JSR INT_DIV 
              AND R3, R3, #0
              LEA R3, ARRAY
              AND R4, R4, #0
              ADD R4, R4, #1
    ARRAY_ADD: STR R4, R3, #0
               AND R0, R0, #0
               NOT R0, R2
               ADD R0, R0, #1
               ADD R0, R0, R4
               BRz #3
               ADD R3, R3, #1
               ADD R4, R4, #1
               BR ARRAY_ADD
           HALT

;;;;;;;;;gets integer number of image-parts to be revealed
INT_DIV: AND R4, R4, #0
         ADD R4, R4, #-13
         AND R3, R3, #0
         ADD R3, R3, R2
         AND R2, R2, #0
    DIVLOOP: ADD R3, R3, R4
             BRn #2
             ADD R2, R2, 1
             BR DIVLOOP
         RET

;;;;;;;;;multiplies progress by 9
NINE_MULT: AND R4, R4, #0
           ADD R4, R4, R2
           AND R3, R3, #0
           ADD R3, R3, #9
           AND R2, R2, 0
    MULTLOOP: ADD R2, R2, R4
              ADD R3, R3, #-1
              BRp MULTLOOP
           RET

;;;;;;;;;calculates progress 
PROG: LD R2, NUM
      NOT R0, R1
      ADD R0, R0, #1
      ADD R2, R2, R0
      RET

ARRAY: .BLKW 9
NUM: .BLKW 1
.END
```
